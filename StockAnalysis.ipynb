```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2017,1,1)
end = datetime.datetime(2021,6,15)
```

```python
#Taking stock data from Datareader
tesla=web.DataReader("TSLA",'yahoo',start,end)
ford=web.DataReader("f",'yahoo',start,end)
gm=web.DataReader("GM",'yahoo',start,end)
```

```python
#saving data in csv formate
tesla.to_csv('Tesla_Stock.csv') 
ford.to_csv('ford_Stock.csv')
gm.to_csv('gm_Stock.csv')
```

```python
#Plotting Open price graph of all 3 stocks together
tesla ['Open'].plot(label='TESLA  Price')
ford ['Open'].plot(label='FORD  Price',figsize=(15,7))
gm ['Open'].plot(label='GM  Price')
plt.legend()
plt.title('Stock Price of Tesla, Ford, GM')
plt.ylabel('Stock price')
plt.show()
```

```python
#Volume Traded by tesla, ford, gm
tesla['Volume'].plot(label='Tesla',figsize=(15,7))
ford['Volume'].plot(label='Ford')
gm['Volume'].plot(label='GM')
plt.title('Volume Traded by Tesla, Ford, GM')
plt.legend()
plt.ylabel('Volume Traded')
```

```python
#DAILY PERCENTAGE CHANGE 
tesla['returns']=(tesla['Close']/tesla['Close'].shift(1))-1
ford['returns']=(ford['Close']/ford['Close'].shift(1))-1
gm['returns']=(gm['Close']/gm['Close'].shift(1))-1
```

```python
#Explaing Volatility with the help of Daily percentage change value
#More flatten the hitoric curve more volatile it is as high standard deviation is dirrectly proptional to high variance

tesla['returns'].hist(bins=100,label='Tesla',alpha=0.5,figsize=(13,6))
ford['returns'].hist(bins=100,label='Ford',alpha=0.5)
gm ['returns'].hist(bins=100,label='Gm',alpha=0.5)
plt.legend()

#plotting unnormalized histo graph
```

```python
#Cumulative Return-
#(what if i invested 1 dollar in this company in past and whats the value today)


tesla['Cumulative Return'] =(1+tesla['returns']).cumprod()
ford['Cumulative Return'] =(1+ford['returns']).cumprod()
gm['Cumulative Return'] =(1+gm['returns']).cumprod()
```

```python
tesla['Cumulative Return'].plot(label='Tesla',figsize=(15,7))
ford['Cumulative Return'].plot(label='Ford')
gm['Cumulative Return'].plot(label='GM')
plt.title('Cumulative Return vs time')
plt.legend()
```
